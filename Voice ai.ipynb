```python
from fastapi import FastAPI, UploadFile
import whisper

app = FastAPI()
model = whisper.load_model("base")

@app.post("/voice-input")
async def voice_input(file: UploadFile):

    with open("audio.mp3", "wb") as f:
        f.write(await file.read())

    result = model.transcribe("audio.mp3")
    text = result["text"]

    return {"transcript": text}
```

```python
from fastapi import FastAPI, UploadFile, File
import whisper
from gtts import gTTS
import shutil

app = FastAPI()

# Load Whisper model once
model = whisper.load_model("base")


def fake_llm_response(text: str):
    """
    Simple placeholder for LLM.
    Replace with OpenAI / Groq later.
    """
    return f"You said: {text}. That sounds interesting!"


@app.post("/voice-input")
async def voice_input(file: UploadFile = File(...)):

    # Save uploaded audio
    with open("input_audio.mp3", "wb") as buffer:
        shutil.copyfileobj(file.file, buffer)

    # 1️⃣ Speech-to-Text
    result = model.transcribe("input_audio.mp3")
    transcript = result["text"]

    # 2️⃣ LLM Processing
    response_text = fake_llm_response(transcript)

    # 3️⃣ Text-to-Speech
    tts = gTTS(response_text)
    tts.save("response_audio.mp3")

    return {
        "transcript": transcript,
        "response_text": response_text,
        "audio_file": "response_audio.mp3"
    }
```
